```python
import os
import boto3
from dotenv import load_dotenv

BUCKET_NAME = "zadmod-9"    ###### Nazwa  bucketu S3 w DigitalOcean

load_dotenv()

s3 = boto3.client("s3")

# Uzyskujemy bieżący katalog
current_folder = os.getcwd()

# Przeszukujemy bieżący katalog
for root, dirs, files in os.walk(current_folder):
    for file in files:
        # Pomijamy pliki z rozszerzeniem .ipynb
        if file.endswith('.ipynb'):
            continue  # przejdź do następnego pliku

        file_path = os.path.join(root, file)
        # Tworzę ścieżkę w S3 bez ścieżki do folderu lokalnego, tylko relatywne
        s3_path = "Dane_mod9/" + os.path.relpath(file_path, current_folder).replace("\\", "/")

        print(f"Wysyłam plik {file_path} do {s3_path}")
        s3.upload_file(
            Filename=file_path,
            Bucket=BUCKET_NAME,
            Key=s3_path
        )
```
